```python
# Import statements
from matplotlib import pyplot
from sklearn import linear_model, metrics, model_selection
import pandas as pd

# Set dataframe to CSV and name columns
names = ["Category", "Displacement (cc)", "Weight (kg)"]
df = pd.read_csv("Bike_Data.csv", names=names)

# Create model
mylog_model = linear_model.LogisticRegression(max_iter=2000, solver='liblinear')
# Assign dependent subset
y = df.values[:, 0]
X = df.values[:, 1:3]

# Train
mylog_model.fit(X, y)

# UI and data entry
weight = input("Please enter the weight of the motorcycle in kg \nEnter Here: ")
weight = int(weight)

displacement = input("\nPlease enter the displacement of the motorcycle in cc\nEnter Here: ")
displacement = int(displacement)

# Prediction output
print("\nThe category of this motorcycle is: " + str(mylog_model.predict([[weight, displacement]])))

# Calculate and display accuracy
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.3)
y_pred = mylog_model.predict(X_test)
print("\nAccuracy: " + str(metrics.accuracy_score(y_test, y_pred) * 100) + "%")


```
